```python
df["type"].value_counts().plot(kind="bar", title="Transaction Types", color="deepskyblue")
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.show()
```

```python
fraud_by_type = df.groupby("type")["isFraud"].mean().sort_values(ascending=False)
fraud_by_type.plot(kind="bar", title="Fraud Rate by Type", color="darkorange")
plt.ylabel("Fraud Rate")
plt.show()
```

```python
df["amount"].describe().astype(int)
```

```python
sns.histplot(np.log1p(df["amount"]), bins=100, kde=True, color="forestgreen")
plt.title("Transaction Amount Distribution (Log Scaled)")
plt.xlabel("Log(Amount + 1)")
plt.show()
```

```python
sns.boxplot(data= df[df["amount"] < 50000], x="isFraud", y="amount", color="darkviolet")
plt.title("Amount vs isFraud (Filtered Under 50k)")
plt.show()
```

```python
df["balanceDiffOrig"] = df["oldbalanceOrg"] - df["newbalanceOrig"]
df["balanceDiffDest"] = df["newbalanceDest"] - df["oldbalanceDest"]
```

```python
frauds_per_step = df[df["isFraud"] == 1]["step"].value_counts().sort_index()
plt.plot(frauds_per_step.index, frauds_per_step.values, label="Fraud per Step", color="red")
plt.xlabel("Step (Time)")
plt.ylabel("Number of Frauds")
plt.title("Frauds Over Time")
plt.grid(True)
plt.show()
```
